```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Crea un objeto Ticker de yfinance con el símbolo de la lista índice S&P500
sp500 = yf.Ticker("^GSPC")
```

```python
# método history del objeto Ticker para obtener los datos históricos del símbolo posterior al año 2000
sp500_data = sp500.history(start="2000-01-01")
```

```python
# Visualizacion inicial de los primeros registros
sp500_data.head()
```

```python
sp500_data.info()
```

```python
sp500_data.describe()
```

```python
# función pairplot de seaborn para visualizar la relación entre las variables numéricas del conjunto de dato
sns.pairplot(sp500_data)
plt.show()
```

La función pairplot crea un gráfico de dispersión para cada par de variables numéricas del conjunto de datos, y un histograma para cada variable numérica en la diagonal

```python
# función heatmap de seaborn para visualizar la matriz de correlación entre las variables numéricas del conjunto de datos
corr = sp500_data.corr()
sns.heatmap(corr, cmap="YlGnBu")
plt.show()
```

La función heatmap crea un mapa de calor que muestra la matriz de correlación entre las variables numéricas del conjunto de datos

```python
# función lineplot de seaborn para visualizar la serie temporal de precios de cierre
sns.lineplot(x="Date", y="Close", data=sp500_data)
plt.show()
```

La función lineplot crea un gráfico de líneas que muestra la serie temporal de precios de cierre. El argumento x es la columna que contiene las fechas, el argumento y es la columna que contiene los precios de cierre y el argumento data es el conjunto de datos

### Graficar outliers y patrones en los datos

```python
# Gráfico de diagrama de caja para el precio de cierre
sns.boxplot(x=sp500_data["Close"])
plt.show()
```

```python
# Gráfico de dispersión para el precio de cierre y el volumen
sns.scatterplot(x=sp500_data["Close"], y=sp500_data["Volume"])
plt.show()
```
